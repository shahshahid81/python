#### Question 1

Alongside this note book, four CSV files are specified (one is in fact a TSV file).

For each file, load it using the CSV module, and find the smallest and largest numbers in the data.

All these files contain just lists of numbers - with the exception of a possible header row

````python
import csv

csv.register_dialect("tsv", delimiter="\t")
csv.register_dialect("custom", delimiter="|", quotechar="-")


def get_value(s):
    try:
        return int(s)
    except ValueError:
        return float(s)


def get_min_max(filename, *, dialect="excel", skip_headers=False):
    with open(filename) as f:
        reader = csv.reader(f, dialect=dialect)
        if skip_headers:
            next(reader)
        min_data = None
        max_data = None
        for data in reader:
            data = [get_value(element) for element in data]
            if min_data is None:
                min_data = data[0]
            if max_data is None:
                max_data = data[0]
            min_data = min(min_data, *data)
            max_data = max(max_data, *data)

    return min_data, max_data


print(get_min_max("file1.csv", skip_headers=True))
print(get_min_max("file2.csv"))
print(get_min_max("file3.tsv", dialect="tsv", skip_headers=True))
print(get_min_max("file4.csv", dialect="custom", skip_headers=True))
````

#### Question 2

Given this data structure consisting of a list of dictionaries, write a function that will write this data out to a file, where the column headers (in the first row) are based on the dictionary keys, and the values are flattened out to one row per dictionary (under the corresponding column header).

Note that not all dictionaries contain all the same keys, nor are the keys necessarily in the same order when present.

For "missing" values, your function should just write an empty string.

For example, given this `data`:

````python
data = [
    {'a': '1_a', 'b': '1_b', 'c': '1_c'},
    {'c': '2_c', 'd': '2_d'},
    {'a': '3_a', 'c': '3_c', 'e': '3_e'}
]
````

```
a,b,c,d,e
1_a,1_b,1_c,,,
,,2_c,2_d,
3_a,,3_c,,3_e
```

The order of the columns and rows is not important - as long as they match up with respective column headers.

````python
import csv

data = [
    {"a": "1_a", "b": "1_b", "c": "1_c"},
    {"c": "2_c", "d": "2_d"},
    {"a": "3_a", "c": "3_c", "e": "3_e"},
]

unique_keys = set()
for row in data:
    unique_keys.update(row)

headers = sorted(list(unique_keys))

rows = []
for row in data:
    row_data = []
    for key in headers:
        row_data.append(row.get(key, ""))
    rows.append(row_data)

with open("result.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(headers)
    for row in rows:
        writer.writerow(row)
````
